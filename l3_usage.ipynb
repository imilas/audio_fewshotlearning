```python
paths = ["few_shot_vfy_zenodo/" + x for x in list(meta["filename"])]
def embed(path):
#     print("doing %s\n"%path)
    audio, sr = sf.read(path)
    embed,ts = openl3.get_audio_embedding(audio[0:sr],sr,input_repr='mel128',embedding_size=512,verbose=False)
    
    return np.concatenate(embed),path
with Pool(10) as p:
    for e,fname in p.imap(embed,paths):
        with open('few_shot_vfy.csv', 'a') as the_file:
            print(len(e))
            the_file.write(fname+","+",".join(np.char.mod('%f', e[0:3072]))+"\n")
```
